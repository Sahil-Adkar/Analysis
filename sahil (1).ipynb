```python
import pandas as pd

# Load the dataset into a pandas DataFrame
df = pd.read_csv("World Happiness Report.csv")

# Check if 'Country Name' column has any duplicate values
has_duplicates = df['Country Name'].duplicated().any()

# Print the result
if has_duplicates:
    print("There are duplicate values in the 'Country Name' column.")
else:
    print("There are no duplicate values in the 'Country Name' column.")
```

```python

# Filter the DataFrame to only include data for the year 2022
df_2022 = df[df['Year'] == 2022]

# Find the country with the highest Life Ladder in 2022
country_with_max_lifeladder = df_2022.loc[df_2022['Life Ladder'].idxmax()]['Country Name']

# Print the result
print("The country with the highest Life Ladder in 2022 is", country_with_max_lifeladder)
```

```python

# Filter the DataFrame to only include data for the year 2022
df_2022 = df[df['Year'] == 2008]

# Sort the DataFrame by 'Life Ladder' score in descending order and select the top 10 rows
top_10_countries = df_2022.sort_values('Life Ladder', ascending=False).head(10)

# Print the result
print("Top 10 countries with the highest Life Ladder score in 2008:")
top_10_countries[['Country Name', 'Life Ladder']]
```

```python
# Calculate the average Life Ladder score of the top 10 countries
avg_lifeladder_top10 = top_10_countries['Life Ladder'].mean()

# Print the result
print("Average Life Ladder score of the top 10 countries in 2022:", avg_lifeladder_top10)
```

```python
# Initialize an empty DataFrame to store the results
result_df = pd.DataFrame(columns=['Year', 'Avg Life Ladder Top 10'])

# Loop through each year from 2006 to 2022
for year in range(2006, 2023):
    # Filter the DataFrame to only include data for the current year
    df_year = df[df['Year'] == year]

    # Sort the DataFrame by 'Life Ladder' score in descending order and select the top 10 rows
    top_10_countries = df_year.sort_values('Life Ladder', ascending=False).head(10)

    # Compute the average Life Ladder score of the top 10 countries
    avg_life_ladder_top_10 = top_10_countries['Life Ladder'].mean()

    # Append the results to the DataFrame
    result_df = result_df.append({'Year': year, 'Avg Life Ladder Top 10': avg_life_ladder_top_10}, ignore_index=True)

#print the result
result_df

```

# The above obeservation shows that avg Happiness in top 10 counties have fallen since 2008 meaning the ecconomy recovered through the crash but Happiness didnt recover


```python
# Define a list of years to iterate over
years = list(range(2006, 2023))

# Loop over each year and check if countries with high GDP have a Life Ladder score above the average for that year
for year in years:
    # Filter the DataFrame to only include data for the current year
    df_year = df[df['Year'] == year]
    
    # Calculate the average Life Ladder score for the current year
    avg_lifeladder = df_year['Life Ladder'].mean()
    
    # Filter the DataFrame to only include data for the current year and countries with high GDP
    df_high_gdp = df_year[df_year['Log GDP Per Capita'] > df_year['Log GDP Per Capita'].mean()]
    
    # Calculate the average Life Ladder score for the countries with high GDP for the current year
    avg_lifeladder_high_gdp = df_high_gdp['Life Ladder'].mean()
    
    # Check if the average Life Ladder score for the countries with high GDP is above the average for all countries for the current year
    if avg_lifeladder_high_gdp > avg_lifeladder:
        print(f"Countries with high GDP had a Life Ladder score above average in {year}.")
    else:
        print(f"Countries with high GDP did not have a Life Ladder score above average in {year}.")
```

# Above observation shows that countries with high gpd have high happiness score

```python
# Define a list of years to iterate over
years = list(range(2010, 2023))

# Loop over each year and check if countries with low GDP per capita have a Life Ladder score below the average for that year
for year in years:
    # Filter the DataFrame to only include data for the current year
    df_year = df[df['Year'] == year]
    
    # Calculate the average Life Ladder score for the current year
    avg_lifeladder = df_year['Life Ladder'].mean()
    
    # Filter the DataFrame to only include data for the current year and countries with low GDP per capita
    df_low_gdp = df_year[df_year['Log GDP Per Capita'] < df_year['Log GDP Per Capita'].mean()]
    
    # Calculate the average Life Ladder score for the countries with low GDP per capita for the current year
    avg_lifeladder_low_gdp = df_low_gdp['Life Ladder'].mean()
    
    # Check if the average Life Ladder score for the countries with low GDP per capita is below the average for all countries for the current year
    if avg_lifeladder_low_gdp < avg_lifeladder:
        print(f"Countries with low GDP per capita had a Life Ladder score below average in {year}.")
    else:
        print(f"Countries with low GDP per capita did not have a Life Ladder score below average in {year}.")
```

# Above obesrvation shows that countries with lower gdp have a happiness index that is lower than average

```python
# Define a list of years to iterate over
years = list(range(2010, 2023))

# Loop over each year and get the top 10 countries with the highest Life Ladder score
for year in years:
    # Filter the DataFrame to only include data for the current year
    df_year = df[df['Year'] == year]
    
    # Sort the DataFrame by Life Ladder score in descending order
    df_year_sorted = df_year.sort_values('Life Ladder', ascending=False)
    
    # Get the top 10 countries with the highest Life Ladder score for the current year
    top_10 = df_year_sorted['Country Name'].head(10).tolist()
    
    # Print the list of top 10 countries for the current year
    print(f"\nTop 10 countries with highest Life Ladder score in {year}:\n")
    for i, country in enumerate(top_10):
        print(f"{i+1}. {country}")
```
